```python

#QUICK SNACK VENDING MACHINE
#Welcoming the user to the vending machine.
print("_____________________________________________")
print("\n")
print("------- QUICK SNACK VENDING MACHINE----------")
print("\n")
print("_____________________________________________\n")

#Def function to greet user
def greet_user():
  print("      Welcome to the Vending Machine!        \n")

greet_user()

#Provides datetime object containing current date and time
from datetime import datetime
now = datetime.now()
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
print("    Date & Time =", dt_string) 
print("\n")


#Starting Vending machine
item1 = {('Doritos','S1') : (3, 5) , ('Lays', 'S2') : (2, 5) , ('Cheetos', 'S3') : (4, 5) , ('Takis', 'S4') : (2, 5) }
item2 = {('Coke', 'D1') : (5, 5) , ('Sprite', 'D2') : (5, 5) , ('Fanta', 'D3') : (8, 5) , ('Milktea', 'D4') : (10, 5) }
item3 = {('M&M', 'C1') : (2, 5) , ('Reese', 'C2') : (3, 5) , ('Dairy Milk', 'C3') : (5, 5) , ('Sneakers', 'C4') : (3, 5) }
item4 = {('Chicken Sandwich', 'F1') : (15, 5) , ('Cheese Sandwich','F2') : (15, 5) , ('Onigiri', 'F3') : (20, 5) , ('Nissin Cup Noodles', 'F4') : (5, 5) }

#Creating list from dictionary of item lists
list = {"S1": "Doritos", "S2": "Lays", "S3": "Cheetos", "S4": "Takis" ,
        "D1": "Coke", "D2": "Sprite", "D3": "Fanta", "D4": "Milktea" ,
        "C1": "M&M", "C2": "Reese", "C3": "Dairy Milk", "C4": "Sneakers" ,
        "F1": "Chicken Sandwich", "F2": "Cheese Sandwich", "F3" : "Onigiri", "F4": "Nissan Cup Noodles"}

#Asking user to start the program using while loop and if-else statements
#Begins interaction with the user
while True:
   start = str(input('Enter "S" to start : ')).upper()
   if start == "S":
    print("\nMENU:")
    
#View items
    for keys, value in item1.items(): #Loop the list of items
     print('(Chips)\t\tName:', keys[0],'---', keys[1]) #Key[0] represents product, Key[1] represents product's code
     print('\t\tPrice:', str(value[0]) + " AED") #Value[0] represents price
     print('\t\tStock:', value[1], '\n') #Value[1] represents stock

    for keys, value in item2.items():
     print('(Drinks)\tName:', keys[0],'---', keys[1])
     print('\t\tPrice:', str(value[0]) + " AED")
     print('\t\tStock:', value[1], '\n')

    for keys, value in item3.items():
     print('(Chocolates)\tName:', keys[0],'---', keys[1])
     print('\t\tPrice:', str(value[0]) + " AED")
     print('\t\tStock:', value[1], '\n')

    for keys, value in item4.items():
     print('(Foods)\t\tName:', keys[0],'---', keys[1])
     print('\t\tPrice:', str(value[0]) + " AED")
     print('\t\tStock:', value[1], '\n')

    print("***Instruction: Find the code NEXT to the name of the item you wish to buy.\n\t\tType in to get the item. (e.g., S1, D2, M3, F4)***\n")
    break
   else:
     print('Try again!') #Error Strings


#Insert money
acceptable_money = [5, 10, 50, 100] #Specific amount of money allowed in the machine

while True:
  insert_money = int(input("Enter The Amount of Money Into The Machine (5, 10, 50, 100):\n"))
  if insert_money in acceptable_money:
    print("You have entered", insert_money, "AED in the machine.\n") #Print the amount of money the user have entered 
    break
  else:
    print("Invalid Amount Entered.\n") #the user input incorrect number #The user will be prompted to retry again
  


amount = insert_money
items = [] # Initialize empty list

#Asking the user to select a product
while True:
  product = input("Select Any Desired Product: ").title()
  if input == "quit":
    start()

#Chips
  elif product == "S1": #User input the code of item
    if amount < 4:
      print("You don't have enough money.\n") #If the insert amount money user put is less than the price, gives a prompt error message
    else:
      print("You have added...\nDoritos") #Informing the user what item they choice 
      amount -= 3  #Deduct the amount the user has entered from the insert money
      break #loop will break when user input valid code
  elif product == 'S2':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nLays")
      amount -= 3
      break
  elif product == 'S3':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nCheetos")
      amount -= 4
      break
  elif product == 'S4':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nTakis")
      amount -= 2
      break
#Drinks
  elif product == 'D1':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added.. \nCoke")
      amount -= 5
      break
  elif product == 'D2':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nSprite")
      amount -= 5
      break
  elif product == 'D3':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nFanta")
      amount -= 8
      break
  elif product == 'D4':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nMilktea")
      amount -= 10
      break
  elif product == 'C1':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nM&M")
      amount -= 2
      break
  elif product == 'C2':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added...\nReese")
      amount -= 3
      break
  elif product == 'C3':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nDairy Milk")
      amount -= 5
      break
  elif product == 'C4':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nSneakers")
      amount -= 3
      break
  elif product == 'F1':
    if amount < 10:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nChicken Sandwich")
      amount -= 15
      break
  elif product == 'F2':
    if amount < 10:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nCheese Sandwich")
      amount -= 15
      break
  elif product == 'F3':
    if amount < 20:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nOnigiri")
      amount -= 20
      break
  elif product == 'F4':
    if amount < 4:
      print("You don't have enough money.\n")
    else:
      print("You have added... \nNissin Cup Noodles")
      amount -= 5
      break

  elif product not in item1 or product not in item2 or product not in item3 or product not in item4: #if user input code not from list inventory
    print("Item not avaiable. Please Try Again.\n") #Print item is not in the vending machine
    continue

items.append(list[product]) #method for showing user's 1st chosen item to the receipt

#Asking the user if they want to purchase another item
while True:
  add = input("\nAny Thing Else?(YES/NO): ").upper()

  if add == "YES":
    product = input("\nSelect Any Product: ").title() #Asking the user to select an item again
#Chips List
    if product == "S1":
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nDoritos")
        amount -= 3
        break
    elif product == 'S2':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nLays")
        amount -= 3
        break
    elif product == 'S3':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nCheetos")
        amount -= 4
        break
    elif product == 'S4':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nTakis")
        amount -= 2
        break
#Drinks List
    elif product == 'D1':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nCoke")
        amount -= 5
        break
    elif product == 'D2':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nSprite")
        amount -= 5
        break
    elif product == 'D3':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nFanta")
        amount -= 8
        break
    elif product == 'D4':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nMilktea")
        amount -= 10
        break
#Chocolates List
    elif product == 'C1':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nM&M")
        amount -= 2
        break
    elif product == 'C2':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nReese")
        amount -= 3
        break
    elif product == 'C3':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nDairy Milk")
        amount -= 5
        break
    elif product == 'C4':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nSneakers")
        amount -= 3
        break
    elif product == 'F1':
      if amount < 15:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nChicken Sandwich")
        amount -= 15
        break
    elif product == 'F2':
      if amount < 15:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nCheese Sandwich")
        amount -= 15
        break
    elif product == 'F3':
      if amount < 20:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nOnigiri")
        amount -= 20
        break
    elif product == 'F4':
      if amount < 4:
        print("You don't have enough money.\n")
      else:
        print("You have added...\nNissin Cup Noodles")
        amount -= 5
        break
      
    elif product not in item1 or product not in item2 or product not in item3 or product not in item4:
      print("Item not available. Please Try Again.")
      continue
    items.append(product)

  elif add == "NO":
    break
  
  else:
    print("Invalid Key Entered. Try Again.") #Prints if incorrect code inputted by the user then loop back to selection 
    continue

items.append(list[product]) #method for showing user's 2nd chosen items to the receipt

#Using Loop to ask user if they want a receipt
while True:
  receipt = input("\nDo you want your receipt?(YES/NO): ").upper()
  
  if receipt == "YES":
    print("\nYOUR ITEMS:") #Prints the items from the append list
    for i in items: #Using loop through a range() to list user's chosen items
      print("- " + i)

    print("\nYOUR MONEY:", str(insert_money) + " AED") #Shows the money entered by the user before
    print("TOTAL AMOUNT:", str(insert_money-amount) + " AED") #Deducting the money inserted from the item's value
    print("YOUR CHANGE:", str(amount) + " AED") #Prints the user's change
    break

  else:
    print("You have a total change of", str(amount) +" AED.") #Prints the user's change
    break

#End Greetings
print("_____________________________________________")
print("\n")
print("              ***THANK YOU!!!***             ")
print("\n")
print("         ***Have a nice day ahead :)***      ")
print("_____________________________________________")
#End

```
